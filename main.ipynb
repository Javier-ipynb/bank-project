```python
# Function to create a violin plot with hue coloring
def violin(x, y, dataframe=df, hue='y'):
    # Set the figure size for the plot
    plt.figure(figsize=(12, 5))
    # Create a violin plot with specified x, y, hue, color palette, and split by hue
    ax = sns.violinplot(data=dataframe, x=x, y=y, hue=hue, palette=['black', 'yellow'], split=True)
    # Rotate x-axis labels if x is a categorical column
    if x in cat_cols:
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
    # Display the plot
    plt.show()

# Function to create a scatter plot with hue coloring
def scat(x, y, dataframe=df, hue='y'):
    # Set the figure size for the plot
    plt.figure(figsize=(12, 7))
    # Create a scatter plot with specified x, y, hue, color palette, and marker style
    sns.scatterplot(data=dataframe, x=x, y=y, hue=hue, palette=['black', 'b'], marker='d')
    # Display the plot
    plt.show()
```

# <p style="background-color:#174262;font-family:newtimeroman;color:#a0c6c6;font-size:150%;text-align:center;border-radius:40px 40px;">Data Preprocessing</p>

```python
# Extract values of the 'campaign' column from the DataFrame
campaign = df['campaign'].values

# Extract values of the 'duration' column from the DataFrame
duration = df['duration'].values

# Extract values of the 'pdays' column from the DataFrame
pdays = df['pdays'].values

# Extract values of the 'previous' column from the DataFrame
previous = df['previous'].values

# Extract values of the 'month' column from the DataFrame
month = df['month'].values

# Extract values of the 'poutcome' column from the DataFrame
poutcome = df['poutcome'].values

# Create an array X with the selected feature columns, transposed
X = np.array([campaign, duration, pdays, previous, month, poutcome]).T

# Extract and ravel the target column values
y = df['y'].values.ravel()

# Split the dataset into training and testing sets with a test size of 20% and a fixed random state
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=75806)
```

# <p style="background-color:#174262;font-family:newtimeroman;color:#a0c6c6;font-size:150%;text-align:center;border-radius:40px 40px;">Model Selection</p>

To proceed, let's create a function where we will use various algorithms to superficially check which one performs best. The reference metric we'll use is f1_score. The main reason for prioritizing this metric is that it ensures the model's performance is balanced and doesn't show any preference during prediction. Given this specific context, we don't want the model to have any bias.

```python
def BasicAlgorithms(X_train, y_train):
    models = {
    'LogisticRegression': LogisticRegression(n_jobs=-1),
    'SVC': SVC(random_state=75806),
    'DecisionTreeClassifier': DecisionTreeClassifier(random_state=75806),
    'RandomForestClassifier': RandomForestClassifier(random_state=75806, n_jobs=-1),
    'GradientBoostingClassifier': GradientBoostingClassifier(random_state=75806),
    'AdaBoostClassifier': AdaBoostClassifier(random_state=75806),
    'KNeighborsClassifier': KNeighborsClassifier(n_jobs=-1),
    'GaussianNB': GaussianNB(),}
    
    model_scores = []
    for name, model in models.items():
    # Perform cross-validation
        cv_scores = cross_val_score(model, X_train, y_train, cv=7, scoring='f1')
    
    # Calculate evaluation metrics
        accuracy = cv_scores.mean()

    # Append the metrics to model_scores
        model_scores.append((name, accuracy))

    # Create a DataFrame from model_scores
    columns = ['Model', 'F1 Score',]
    df_model_scores = pd.DataFrame(model_scores, columns=columns)

    # Sort the DataFrame by F1 values in descending order
    df_model_scores = df_model_scores.sort_values(by='F1 Score', ascending=False)

    # Display the DataFrame
    print(df_model_scores)
    
BasicAlgorithms(X_train, y_train) 
```

Having these values means that the model has some kind of preference when making predictions. Let's remember that the classes are unbalanced, which is very important and often directly impacts performance. It is most likely that due to this imbalance, we get this result in the metric, which makes our model not balanced at all. Although we could calculate other metrics like accuracy, which basically gives us the percentage of correct predictions, we would be making a mistake. Because in a real environment, the model would not be making a fair prediction, but rather has a preference when doing so. And this could lead to constant errors in real environments. So we will try to level the classes by undersampling, that is, by eliminating data from the class with the most data to balance it with the one that has the least. In this way, we can run the previous function again and see if the model has improved or not.

```python
# Create a DataFrame with rows where 'y' is 0 (negative cases)
df_negative = df[df['y'] == 0]

# Create a DataFrame with rows where 'y' is 1 (positive cases)
df_positive = df[df['y'] == 1]

# Randomly sample 3676 rows from the negative cases DataFrame without replacement
sub_df_negative = df_negative.sample(n=3676, replace=False, random_state=8364)

# Concatenate the positive cases DataFrame with the sampled negative cases DataFrame
sub_sampling = pd.concat([df_positive, sub_df_negative])

# Set the figure size for the plot
plt.figure(figsize=(8,5))

# Create a histogram plot of the 'y' column in the subsampled DataFrame with the color violet
sns.histplot(data=sub_sampling, x='y', color='violet')
```

Now the target variable is balanced. Let's remember that initially the values were: (`no: 29238 (0), yes: 3712 (1)`). Now the class is clearly not unbalanced, so we will run the function again with the different algorithms to see if we observe any improvement compared to the previous metric.

```python
# Extract values of the 'campaign' column from the subsampled DataFrame
campaign = sub_sampling['campaign'].values

# Extract values of the 'duration' column from the subsampled DataFrame
duration = sub_sampling['duration'].values

# Extract values of the 'pdays' column from the subsampled DataFrame
pdays = sub_sampling['pdays'].values

# Extract values of the 'previous' column from the subsampled DataFrame
previous = sub_sampling['previous'].values

# Extract values of the 'month' column from the subsampled DataFrame
month = sub_sampling['month'].values

# Extract values of the 'poutcome' column from the subsampled DataFrame
poutcome = sub_sampling['poutcome'].values

# Create an array X with the selected feature columns, transposed
X = np.array([campaign, duration, pdays, previous, month, poutcome]).T

# Extract and ravel the target column values
y = sub_sampling['y'].values.ravel()

# Split the subsampled dataset into training and testing sets with a test size of 20%, fixed random state, and stratification by target values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=75806, stratify=y)
```

Finally, we can clearly see how we obtain a better metric. Despite being a single metric and very general, this makes it very clear that the model is much more balanced than before, all thanks to working with balanced classes. So, I proceed to choose two of the algorithms with the best scores to train them a bit further and obtain more specific metrics.

```python
# Define the RandomForestModel function
def RandomForestModel(X_train, y_train, X_test, y_test):
    # Initialize the RandomForestClassifier with a fixed random state
    rf = RandomForestClassifier(random_state=75806)

    # Define the parameter grid for GridSearchCV
    param_grid = {
        'n_estimators': [800, 1000, 1200],
        'max_depth': [10, 12, 15],
        'min_samples_split': [13, 15, 18]}
    
    # Initialize GridSearchCV with the parameter grid, 7-fold cross-validation, and parallel processing
    rf = GridSearchCV(estimator=rf, param_grid=param_grid, cv=7, n_jobs=-1)

    # Fit the model to the training data
    rf.fit(X_train, y_train)

    # Print the best hyperparameters found by GridSearchCV
    print(rf.best_params_)

    # Retrieve the best estimator (model) found by GridSearchCV
    rf = rf.best_estimator_

    # Predict the labels of the test set
    y_pred = rf.predict(X_test)

    # Evaluate general metrics of the model's performance
    GeneralMetrics(y_test, y_pred)
    
    # Return the fitted RandomForest model
    return rf
```

```python
# Define the GradientBoostingModel function
def GradientBoostingModel(X_train, y_train, X_test, y_test):
    # Initialize the GradientBoostingClassifier with a fixed random state
    gbr = GradientBoostingClassifier(random_state=75806)

    # Define the parameter grid for GridSearchCV
    param_grid = {
        'n_estimators': [200, 300],
        'max_depth': [3, 5],
        'min_samples_split': [3, 5]}
    
    # Initialize GridSearchCV with the parameter grid, 7-fold cross-validation, and parallel processing
    gbr = GridSearchCV(estimator=gbr, param_grid=param_grid, cv=7, n_jobs=-1)

    # Fit the model to the training data
    gbr.fit(X_train, y_train)

    # Print the best hyperparameters found by GridSearchCV
    print(gbr.best_params_)

    # Retrieve the best estimator (model) found by GridSearchCV
    gbr = gbr.best_estimator_

    # Predict the labels of the test set
    y_pred = gbr.predict(X_test)

    # Evaluate general metrics of the model's performance
    GeneralMetrics(y_test, y_pred)

    # Return the fitted GradientBoosting model
    return gbr
```

The final performance of the model is similar in both cases; we observe that there is no significant improvement between them, and the final decision will depend on the overall context and preferences regarding the preferred algorithm.
